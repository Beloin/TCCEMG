```python
import pandas as pd
import seaborn as sbn
import matplotlib.pyplot as plt

import matplotlib as mpl

from scipy.fftpack import fft, ifft
```

```python
relaxed = pd.read_csv("./emg_hand_raw_data/ponto 1a/relaxado/tek0000ALL.csv", skiprows=20)
contracted = pd.read_csv("./emg_hand_raw_data/ponto 1a/tek0001ALL.csv", skiprows=20)
```

```python
relaxed_fft = fft(relaxed.CH1.values[:200])
relaxed_fft[:10]
```

```python
fig, ax = plt.subplots(figsize=(30, 6))

ax.set_ylim((min(relaxed_fft)-1, max(relaxed_fft)+1))

sbn.lineplot(relaxed_fft)

## TODO: Maybe timestamp is not uniform? Check for uniformity
```
